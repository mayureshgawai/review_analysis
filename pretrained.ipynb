```python
scores = {'correct': 0, 'incorrect': 0}
sent = ''

for i in range(0, len(df['Reviews'])):
    text = df['Reviews'][i]

    if(isinstance(df['Reviews'][i], str)):
        doc = nlp(text)
        score = doc.cats

        if(score['positive'] > score['negative']):
            sent = 'pos'
        elif(score['positive'] < score['negative']):
            sent = 'neg'

        if(sent == df['Sentiment'][i]):
            scores['correct'] += 1
        else:
            scores['incorrect'] += 1


    
print(scores)
```
